### 3. Filter Users and Books

Remove users with **less than 200 ratings** and books with **less than 100 ratings**. This step ensures that only frequent users and popular books are considered, making the recommendations more reliable.

```python
# Get users with >=200 ratings in the original dataset
user_counts = df_ratings['user'].value_counts()
active_users = user_counts[user_counts >= 200].index

# Get books with >=100 ratings in the original dataset
book_counts = df_ratings['isbn'].value_counts()
popular_books = book_counts[book_counts >= 100].index

# Keep only ratings where both conditions hold
df_ratings_filtered = df_ratings[
    df_ratings['user'].isin(active_users) & 
    df_ratings['isbn'].isin(popular_books)
]

print("Filtered ratings shape:", df_ratings_filtered.shape)
print("Number of unique books left:", df_ratings_filtered['isbn'].nunique())
print("Number of unique users left:", df_ratings_filtered['user'].nunique())
```

### 4. Create a User‑Item Matrix

We pivot the filtered ratings so that **rows = books (ISBN)**, **columns = users**, and values are the ratings.  
Empty cells are filled with `0` (meaning the user did not rate that book). This matrix is then converted to a sparse format for memory efficiency.

```python
# Pivot table: books (index) vs users (columns)
pivot = df_ratings_filtered.pivot_table(
    index='isbn',
    columns='user',
    values='rating'
).fillna(0)

# Convert to sparse matrix (required by sklearn)
book_matrix = csr_matrix(pivot.values)

print("Pivot table shape:", pivot.shape)
print("Sparse matrix shape:", book_matrix.shape)
```

### 7. Implementing the Recommendation Function

The function `get_recommends(book)` should:
1. Find the ISBN of the given book title (case‑insensitive match).
2. Locate its position in the pivot table.
3. Ask the KNN model for the **6 nearest neighbours** (including itself).
4. Skip the first neighbour (the book itself) and take the next five.
5. Retrieve their titles and distances.
6. **Sort the five recommendations by distance descending** (as shown in the example).
7. Return a list in the required format: `[book_title, [[title1, dist1], [title2, dist2], ...]]`.

```python
def get_recommends(book=""):
    # ---- 1. Find ISBN for the given title ----
    # Case‑insensitive search (take the first match if multiple)
    book_row = df_books[df_books['title'].str.lower() == book.lower()]
    if book_row.empty:
        print(f"Book '{book}' not found in the dataset.")
        return [book, []]
    
    isbn = book_row.iloc[0]['isbn']
    
    # ---- 2. Check if the book exists in filtered pivot ----
    if isbn not in pivot.index:
        print(f"Book '{book}' is not in the filtered dataset (might have <100 ratings).")
        return [book, []]
    
    # Get index of this book in the pivot table
    book_idx = pivot.index.get_loc(isbn)
    book_vector = book_matrix[book_idx]
    
    # ---- 3. Find nearest neighbours (6 to include itself) ----
    distances, indices = model.kneighbors(book_vector, n_neighbors=6)
    
    # ---- 4. Skip the first (itself) ----
    neighbor_indices = indices.flatten()[1:]   # indices of neighbours in pivot
    neighbor_distances = distances.flatten()[1:]
    
    # ---- 5. Get titles for neighbours ----
    neighbor_isbns = pivot.index[neighbor_indices]
    neighbor_titles = [isbn_to_title.get(isbn, "Unknown") for isbn in neighbor_isbns]
    
    # ---- 6. Combine titles and distances, then sort by distance descending ----
    recommendations = list(zip(neighbor_titles, neighbor_distances))
    recommendations.sort(key=lambda x: x[1], reverse=True)
    
    # ---- 7. Return required format ----
    return [book, recommendations]
```

### 8. Test the Function

The project provides a test function. We run it to check our implementation.

```python
def test_book_recommendation():
    test_pass = True
    recommends = get_recommends("Where the Heart Is (Oprah's Book Club (Paperback))")
    if recommends[0] != "Where the Heart Is (Oprah's Book Club (Paperback))":
        test_pass = False
    recommended_books = ["I'll Be Seeing You", 'The Weight of Water', 'The Surgeon', 'I Know This Much Is True']
    recommended_books_dist = [0.8, 0.77, 0.77, 0.77]
    for i in range(2):
        if recommends[1][i][0] not in recommended_books:
            test_pass = False
        if abs(recommends[1][i][1] - recommended_books_dist[i]) >= 0.05:
            test_pass = False
    if test_pass:
        print("You passed the challenge! 🎉🎉🎉🎉🎉")
    else:
        print("You haven't passed yet. Keep trying!")

test_book_recommendation()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set professional style
sns.set_style("whitegrid")
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.linewidth'] = 1.5
plt.rcParams['xtick.major.width'] = 1.5
plt.rcParams['ytick.major.width'] = 1.5

def plot_recommendations(book_title, save_path='figure.png'):
    """
    Generate a bar chart of the top 5 recommended books with distances.
    """
    recs = get_recommends(book_title)
    if not recs[1]:  # if no recommendations
        print(f"No recommendations found for '{book_title}'")
        return

    # Extract titles and distances
    titles = [item[0] for item in recs[1]]
    distances = [item[1] for item in recs[1]]

    # Create figure and axis
    fig, ax = plt.subplots(figsize=(10, 6))

    # Use a colormap (darker = smaller distance / more similar)
    colors = plt.cm.Blues(0.3 + 0.5 * (1 - np.array(distances)/max(distances)))

    # Horizontal bar chart (top to bottom, highest distance first)
    bars = ax.barh(titles, distances, color=colors, edgecolor='black', linewidth=1.2)

    # Add distance values on bars
    for bar, dist in zip(bars, distances):
        ax.text(bar.get_width() - 0.02, bar.get_y() + bar.get_height()/2,
                f'{dist:.4f}', va='center', ha='right', fontsize=11, fontweight='bold', color='white')

    # Labels and title
    ax.set_xlabel('Cosine Distance (lower = more similar)', fontsize=13, fontweight='semibold')
    ax.set_title(f'Top 5 Books Similar to\n"{book_title}"', fontsize=16, fontweight='bold', pad=20)
    ax.invert_yaxis()  # highest distance on top (optional, you can remove if you want highest similarity on top)

    # Adjust spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_linewidth(1.5)
    ax.spines['left'].set_linewidth(1.5)

    # Add a light background grid
    ax.xaxis.grid(True, linestyle='--', alpha=0.7)
    ax.set_axisbelow(True)

    plt.tight_layout()
    save_path = book_title.replace(' ','_').replace(')','_').replace("'","_").replace('(','_')
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()
    #print(f"Figure saved as '{save_path}'")

plot_recommendations("Where the Heart Is (Oprah's Book Club (Paperback))")
```
